Genera una visualizaciÃ³n que muestre las principales medidas estadÃ­sticas de cada columna

```python
mlb_frames.columns
```

```python
percentil_25 = mlb_frames[' Age'].quantile(0.25)
percentil_50 = mlb_frames[' Age'].quantile(0.5)
percentil_75 = mlb_frames[' Age'].quantile(0.75)
```

```python
sns.histplot(mlb_frames[' Age'], bins=40,)
media = mlb_frames[' Age'].mean()
plt.axvline(media, color='orange', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(percentil_25, color='r', linestyle='dashed', linewidth=1, label='25th Percentile')
plt.axvline(percentil_50, color='g', linestyle='dashed', linewidth=1, label='50th Percentile (Median)')
plt.axvline(percentil_75, color='b', linestyle='dashed', linewidth=1, label='75th Percentile')
plt.title('Histograma para la Edad')
plt.xlabel("Chupar")
plt.legend()


```

```python
sns.kdeplot(mlb_frames[' Age'], fill=True)
```

```python
percentil_25 = mlb_frames[' Height(inches)'].quantile(0.25)
percentil_50 = mlb_frames[' Height(inches)'].quantile(0.5)
percentil_75 = mlb_frames[' Height(inches)'].quantile(0.75)
sns.histplot(mlb_frames[' Height(inches)'], bins=20,)
media = mlb_frames[' Height(inches)'].mean()
plt.axvline(media, color='orange', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(percentil_25, color='r', linestyle='dashed', linewidth=1, label='25th Percentile')
plt.axvline(percentil_50, color='g', linestyle='dashed', linewidth=1, label='50th Percentile (Median)')
plt.axvline(percentil_75, color='b', linestyle='dashed', linewidth=1, label='75th Percentile')
plt.title('Histograma para la Altura')
plt.xlabel("Alturar")
plt.legend()
```
